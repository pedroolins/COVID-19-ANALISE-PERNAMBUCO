# Variáveis do Conjunto de dados
### Dicionário das variáveis
* **regiao:** Indica a região do Brasil onde foi regsitrada a observação. **Note que onde temos "Brasil" na variável região, são os dados gerais do país somando os dados de todas as regiões.**
* **estado:** Indica o estado do Brasil onde foi regsitrada a observação. **Note que quando o estado for um NaN trata-se dos dados gerais de cada região do país, somando todos os dados dos estados das determinadas regiões naquela data registrada.**
* **municipio:** Indica o município do Brasil onde foi regsitrada a observação. **Note que quando o município for um NaN e a variável codmun também for um NaN, trata-se dos dados gerais do determinado estado no qual estamos trabalhando, somando todos os dados de cada município do estado naquela data registrada.**
* **coduf:** Indica o código da Unidade Federativa onde foi registrada a observação.
* **codmun:** Indica o código do município onde foi registrada a observação.
* **codRegiaoSaude:** Indica o código da região de saúde onde foi registrada a observação.
* **NomeRegiaoSaude:** Indica o nome da região de saúde onde foi registrada a observação.
* **data:** Indica a data em que foi registrada a observação.
* **semanaEpi:** Indica a semana epidemiólgica em que foi registrada a observação.
* **populacaoTCU2019:** Indica a população estimada de 2019 utilizada pelo TCU de onde foi registrada a observação. Para saber mais acesse [aqui.](https://datasus.saude.gov.br/populacao-residente/)
* **casosAcumulados:** Número total de casos confirmados por COVID-19 que foram registrados pelas Secretarias Municipais e Estaduais de Saúde no período considerado.
* **casosNovos:** Número de casos novos confirmados por COVID-19 que foram registrados pelas Secretarias Municipais e Estaduais de Saúde em relação ao dia anterior.
* **obitosAcumulado:** Número total de óbitos confirmados por COVID-19 que foram registrados pelas Secretarias Municipais e Estaduais de Saúde no período considerado.
* **obitosNovos:** Número de óbitos novos confirmados por COVID-19 que foram registrados pelas Secretarias Municipais e Estaduais de Saúde em relação ao dia anterior.
* **Recuperadosnovos:** Número de recuperados da COVID-19 que foram registrados pelas Secretarias Municipais e Estaduais de Saúde em relação ao dia anterior.
* **emAcompanhamentoNovos:** São considerados como “em acompanhamento” todos os casos notificados nos últimos 14 dias pelas Secretarias Estaduais de Saúde e que não evoluíram para óbito.
* **interior/metropolitana:** Indica onde a observação foi registrada na região do estado, se no interior (valor 0) ou na região metropolitana (valor 1).

# 8. Análise e visualização dos dados da COVID-19 no estado de Pernambuco

```python
# criando função que gera um gráfico de linha
def plota_linha(df, x, y, xlabel, ylabel, hue, titulo):
    sns.set_style('darkgrid')
    sns.set_palette('dark:salmon_r')
    ax = sns.lineplot(data = df, x = x, y = y, hue = hue)
    ax.figure.set_size_inches(18,6)
    ax.set_title(titulo, fontsize = 18, loc = 'left')
    ax.set_xlabel(xlabel, fontsize = 14)
    ax.set_ylabel(ylabel, fontsize = 14)
```

```python
# criando função que gera gráfico de barras
def plota_barra(df, x, y, titulo, xlabel, ylabel, color = 'red'):
    plt.figure(figsize=(20,6))
    plt.bar(df[x], df[y], color = color)
    plt.title(titulo, loc ='left', fontsize = 18)
    plt.xlabel(xlabel, fontsize = 18)
    plt.ylabel(ylabel, fontsize = 18)
```

```python
# gráfico de casos acumulados em Pernambuco
plota_linha(df = df_pe_geral, x = 'data', y = 'casosAcumulado', hue = None,
            titulo = 'N° de registros de casos Acumulados em Pernambuco', xlabel = 'Datas', ylabel = 'Casos Acumulados')
```

```python
# gráfico de casos por dia em Pernambuco
plota_linha(df = df_pe_geral, x = 'data', y = 'casosNovos', hue = None,
            titulo = 'N° de registros de casos por dia em Pernambuco', xlabel = 'Datas', ylabel = 'N° de Casos')
```

Para contornar essa sazonalidade semanal e gerar uma visualização mais suave irei realizar uma média móvel de 7 dias.</span>

## 8.1. Obtendo a média móvel de 7 dias dos casos e óbitos registrados e dividindo os dados dos dois DataFrames em 2020 e 2021

```python
# criando coluna ano, mês e dia da semana nos DataFrames 
df_pe_geral['Ano'] = df_pe_geral['data'].dt.year
df_pe_geral['mes'] = df_pe_geral['data'].dt.month
df_pe_geral['dia_da_semana'] = df_pe_geral['data'].dt.day_name()
df_pe_geral.head(2)
```

```python
# traduzindo os valores na coluna dia da semana
dia_da_semana = {'Thursday': 'Quinta', 'Friday': 'Sexta', 'Saturday': 'Sabado', 'Sunday': 'Domingo', 'Monday': 'Segunda',
               'Tuesday': 'Terça','Wednesday': 'Quarta'}
df_pe_geral['dia_da_semana'] = df_pe_geral['dia_da_semana'].map(dia_da_semana)
df_pe_geral.head(2)
```

```python
# tabela com o número médio de casos e óbitos registrados por dia da semana durante toda pandemia
df_dia_da_semana = df_pe_geral.groupby('dia_da_semana', as_index = False)[['casosNovos', 'obitosNovos']].mean().round()
df_dia_da_semana.casosNovos = df_dia_da_semana.casosNovos.astype('int64')
df_dia_da_semana.obitosNovos = df_dia_da_semana.obitosNovos.astype('int64')
df_dia_da_semana.sort_values('casosNovos', inplace = True)
df_dia_da_semana.index = range(len(df_dia_da_semana))
df_dia_da_semana
```

### <span style="color: gray">Pela tabela acima fica ainda mais evidente uma sazonalidade semanal nos dados do estado, isto devido a diminuição da divulgação desses dados nos finais de semana. É perceptível que os dias mais afetados com isso são justamente o domingo e a segunda-feira. Portanto, faz-se necessário a utilização de uma média móvel de 7 dias.</span>

```python
# criando as variáveis média móvel de casos e óbitos no dataframe
df_pe_geral['média_móvel_casos'] = df_pe_geral.casosNovos.rolling(window = 7, center = False).mean()
df_pe_geral['média_móvel_óbitos'] = df_pe_geral.obitosNovos.rolling(window = 7, center = False).mean()
```

```python
# visualizando a média móvel de casos
plota_linha(df = df_pe_geral, x = 'data', y = 'média_móvel_casos', hue = None,
            xlabel = 'datas', ylabel = 'N° de casos', titulo = 'Média móvel (7 dias) de casos de covid-19 no estado de Pernambuco')
plt.bar(df_pe_geral['data'], height = df_pe_geral['casosNovos'], color = 'white');
```

```python
# visualizando a média móvel de óbitos
plota_linha(df = df_pe_geral, x = 'data', y = 'média_móvel_óbitos', hue = None,
            xlabel = 'datas', ylabel = 'N° de óbitos',
            titulo = 'Média móvel (7 dias) de óbitos por covid-19 no estado de Pernambuco')
plt.bar(df_pe_geral['data'], height = df_pe_geral['obitosNovos'], color = 'white');
```

```python
# realizando o mesmo procedimento de criar as variáveis 'Ano', 'mes' e 'dia_da_semana' para o DataFrame df_pe_mun
df_pe_mun['Ano'] = df_pe_mun['data'].dt.year
df_pe_mun['mes'] = df_pe_mun['data'].dt.month
df_pe_mun['dia_da_semana'] = df_pe_mun['data'].dt.day_name()
df_pe_mun['dia_da_semana'] = df_pe_mun['dia_da_semana'].map(dia_da_semana)
```

```python
# dividindo os dados dos DataFrames em 2020 e 2021
df_pe_geral_2020 = df_pe_geral[df_pe_geral['Ano'] == 2020]
df_pe_geral_2021 = df_pe_geral[df_pe_geral['Ano'] == 2021]
df_pe_mun_2020 = df_pe_mun[df_pe_mun['Ano'] == 2020]
df_pe_mun_2021 = df_pe_mun[df_pe_mun['Ano'] == 2021]
```

## 8.6. Qual foi o mês com mais registros de casos confirmados por COVID-19 no estado de Pernambuco em 2020 e 2021?

```python
# analisando os meses de 2020
meses_pe_2020 = df_pe_geral_2020.groupby('mes', as_index = False)[['casosNovos', 'obitosNovos']].sum()
meses = {2: 'fevereiro', 3 : 'março', 4: 'abril', 5: 'maio', 6: 'junho',
         7: 'julho', 8: 'aqgosto', 9: 'setembro', 10: 'outubro',
         11: 'novembro', 12: 'dezembro'}
meses_pe_2020.mes = meses_pe_2020.mes.map(meses)
meses_pe_2020.rename(columns = {'mes': 'meses'}, inplace = True)
meses_pe_2020
```

```python
plota_barra(meses_pe_2020, x = 'meses', y = 'casosNovos',
            titulo = 'Casos confirmados de covid-19 por mês em Pernambuco (2020)',
            xlabel = 'meses', ylabel = 'N° Casos confirmados')
```

### <span style="color:gray">É visível que o mês em 2020 com mais casos confirmados foi o de dezembro, com um total de 39.760 casos confirmados no estado, também é importante observar que o segundo mês com mais casos registrados foi o de julho com um total de 36.147 casos confirmados. ### <span style="color: gray">Através do gráfico acima vemos que o número total de óbitos registrados no interior e na região metropolitana do estado, em relação ao número total de casos confirmados, se apresenta de maneira diferente. A região metropolitana apresenta maior número de óbitos do que o interior do estado, e isto deve ocorrer porque quando os hospitais do interior ficam lotados as pessoas em estado crítico são enviadas para os hospitais de referência da região metropolitana, que são maiores e apresentam maior número de leitos. Sendo assim, as pessoas, infelizmente, acabam falecendo nesses hospitais, fazendo com que o número de óbitos na região metropolitana seja maior.<span>

```python
# dataframe dos municípios da região metropolitana com total de óbitos e casos confirmados em 2020
df_regiao_metropolitana = df_pe_mun_2020[df_pe_mun_2020['interior/metropolitana'] == 1][['municipio','obitosAcumulado', 'casosAcumulado']]
df_regiao_metropolitana_agrupado = df_regiao_metropolitana.groupby(['municipio'], as_index = False).max()
df_regiao_metropolitana_agrupado.rename(columns = {'municipio': 'município', 'obitosAcumulado': 'total_de_óbitos',
                                         'casosAcumulado': 'total_de_casos'}, inplace = True)
df_regiao_metropolitana_agrupado.sort_values(['total_de_casos', 'total_de_óbitos'], ascending = False, inplace = True)
df_regiao_metropolitana_agrupado.index = range(len(df_regiao_metropolitana_agrupado))
df_regiao_metropolitana_agrupado.total_de_casos = df_regiao_metropolitana_agrupado.total_de_casos.astype('int64')
df_regiao_metropolitana_agrupado
```

```python
# criando função que plota um gráfico de barras horizontal
def plota_barra_h(df, x, y, titulo, xlabel, ylabel, color = 'red'):
    plt.figure(figsize=(20,10))
    plt.barh(df[y], df[x], color = color)
    plt.title(titulo, loc ='left', fontsize = 18)
    plt.xlabel(xlabel, fontsize = 18)
    plt.ylabel(ylabel, fontsize = 18)
```

```python
# Gráfico de barras das cidades da região metropolitana a cerca do total de óbitos e casos em 2020
plota_barra_h(df = df_regiao_metropolitana_agrupado, x = 'total_de_casos', y = 'município',
            ylabel = 'Municípios', xlabel = 'Total de casos confirmados',
            titulo = 'Total de casos por Município da região metropolitana no ano de 2020')
plota_barra_h(df = df_regiao_metropolitana_agrupado, x = 'total_de_óbitos', y = 'município',
            ylabel = 'Municípios', xlabel = 'Total de óbitos registrados',
            titulo = 'Total de óbitos por Município da região metropolitana no ano de 2020',
            color = 'black')
```

### <span style="color: gray">Utilizando os critérios de total de casos confirmados em 2020, a cidade da região metropolitana mais afetada foi a capital Recife, o que não surpreende muito, pois é a maior cidade do estado de Pernambuco, apresentando a maior população.<span>

### 9.1 A Pandemia vem decrescendo ou aumentando no estado?

### <span style="color: gray">Através de todas as análises realizadas nesse projeto fica evidente que no ano de 2020 a evolução do vírus no estado teve um ápice no meses de maio, junho e julho, e logo após, nos meses de agosto, setembro e outubro houve um decrescimento, tanto no número de casos como também no número de óbitos, Porém isto não permaneceu nos meses seguintes. Hoje, o estado de Pernambuco se encontra em uma situação bastante delicada, com cada vez mais o número de casos e óbitos aumentando. Portanto, a pandemia vem aumentando no estado, e com isso faz-se necessário a utiliazação de medidas mais rígidas para o controle da evolução do vírus, tanto no interior do estado como na região metropolitana, para que desta maneira tenhamos uma situação mais controlada.<span>
